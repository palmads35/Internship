```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants')
```

```python
page
```

```python
# response >500 we cannot use that webpage
```

```python
soup= BeautifulSoup(page.content)
soup
```

```python
first_title= soup.find('div', class_ = "restnt-info cursor")
```

```python
first_title
```

```python
first_title.text
```

```python
loc = soup.find('div', class_ = "restnt-loc ellipsis")
loc
```

```python
loc.text
```

```python
price= soup.find('span', class_ = "double-line-ellipsis")
price.text.replace('₹','').split()[0]
```

```python
titles= []
for i in soup.find_all('div', class_= "restnt-info cursor"):
    titles.append(i.text)
titles
```

```python
loc=[]
for i in soup.find_all('div', class_= "restnt-loc ellipsis"):
    loc.append(i.text)
loc
```

```python
price=[]
for i in soup.find_all('span', class_= "double-line-ellipsis"):
    price.append(i.text)
price
```

```python
images = []
for i in soup.find_all("img", class_="no-img"):
    images.append(i['data-src'])
images
```

```python
print(len(titles),len(loc),len(price),len(images))
```

```python
import pandas as pd
df = pd.DataFrame({'Tiltes':titles,'Location':loc,'Price':price,'Images_url':images})
df
```
