```python
from bs4 import BeautifulSoup
import requests


#https://www.imdb.com/india/top-rated-indian-movies/
```

```python
page = requests.get('https://www.imdb.com/india/top-rated-indian-movies')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
name_title= soup.find('td', class_ = "titleColumn")
name_title
      
```

```python
name_title.text
```

```python
rating= soup.find('td', class_= "ratingColumn imdbRating")
rating
```

```python
rating.text
```

```python
Release =soup.find('span', class_= "secondaryInfo")
Release
```

```python
Release.text
```

```python
titles= []
for i in soup.find_all('td', class_= "titleColumn"):
    titles.append(i.text)
titles
```

```python
rating= []
for i in soup.find_all('td', class_= "ratingColumn imdbRating"):
    rating.append(i.text)
rating
```

```python
release= []
for i in soup.find_all('span', class_= "secondaryInfo"):
    release.append(i.text)
release
```

```python
import pandas as pd
df= pd.DataFrame({'Titles':titles,'Rating':rating,'Year':release})
df.head(10)
```

```python
# Python program to display list of respected former president of India 
page = requests.get('https://presidentofindia.nic.in/former-presidents.htm')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
Former_president= soup.find('div', class_ = "presidentListing")
Former_president
```

```python
Former_president.text.split()[0:3]
```

```python
Term= soup.find('span', class_="terms")
Term
```

```python
Term.text
```

```python
president= []
for i in soup.find_all('div', class_= "presidentListing"):
    president.append(i.text)
president
```

```python
Term= []
for i in soup.find_all('span', class_= "terms"):
    Term.append(i.text)
Term
```

```python
import pandas as pd
df= pd.DataFrame({'Former_President':president,'Term':Term})
df
```
